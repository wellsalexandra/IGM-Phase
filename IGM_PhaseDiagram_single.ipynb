```python
### The following code is used to create a single phase diagram plot,
### given the scale factor and simulation box:

scaleFactor = 0.1609 # Selected from provided simulation data options
simulationBox = 'B' # Selected from options A,B,C,F
arr = np.load("drive/MyDrive/UoM REU Project 2023/igm_phase_data/" + simulationBox + "/nT.a=" + str(scaleFactor)+".bin.npy")

plt.imshow(np.log(arr), extent = [-2, 3, 0, 6], origin = 'lower')
cb = plt.colorbar()
cb.set_label(label="log10(Probabilty)", fontsize='xx-large')

# x-axis is log density, -2 to 3 (inversed cm cubed)
# y-axis is log Temperature, 2 to 8 (kelvin)
plt.xlabel("log10(1+Î´) (/cm^3)", fontsize='xx-large')
plt.ylabel("log10(T/K)", fontsize='xx-large')
plt.title("Phase diagram starter plot", fontsize='xx-large')

```
